```python
import pandas as pd
dataset=pd.read_csv('/bestsellers with categories.csv')
```

```python
dataset.head()
```

The task here is to classify the book (Fiction/Non-Fiction) based on the other columns. I've used simple neural networks in order to classify.

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()

dataset['Name']=encoder.fit_transform(dataset['Name'])

dataset['Author']=encoder.fit_transform(dataset['Author'])

dataset['Genre']=encoder.fit_transform(dataset['Genre'])

dataset.head()
```

```python
y=dataset.Genre 
x=dataset.drop(['Genre'],axis=1)


from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.20)
```

```python
X_train.shape 
```

```python
Y_train.value_counts()  
```

#Neural_Networks

```python
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()

#Input layer
model.add(Dense(8,input_dim=6,kernel_initializer='uniform',activation='relu'))

#Two hidden layers
model.add(Dense(4,input_dim=6,kernel_initializer='uniform',activation='relu'))
model.add(Dense(2,input_dim=6,kernel_initializer='uniform',activation='relu'))

#Output layer
model.add(Dense(1,input_dim=6,kernel_initializer='uniform',activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 

model.fit(X_train, Y_train, epochs=150, batch_size=10) 

y_pred=model.predict(X_test)


from sklearn.metrics import accuracy_score

ac = accuracy_score(Y_test, y_pred.round())
print('accuracy of the model: ', ac)
```
